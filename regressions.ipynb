```python
import polars as pl

# month order
month_order = ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"]

data = pl.read_csv("data.csv", ignore_errors=True)
data
```

```python
# show distinct
data["reporte_descripcion"].unique()
```

```python
# get all that have months in the "cohorte" column
data_month = data.filter(pl.col("cohorte").str.contains(r"\b(?:Enero|Febrero|Marzo|Abril|Mayo|Junio|Julio|Agosto|Septiembre|Octubre|Noviembre|Diciembre)\b"))

# add time column
data_month = data_month.with_columns(
    date = pl.date(pl.col("año"), pl.col("cohorte").replace(month_order, range(1, 13)), 1)
)


data_month
```

```python
# scatterplot for all data_month
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures, MinMaxScaler

A = data_month.filter(
    pl.col("reporte_descripcion").str.contains("Ingresos Programa de Salud Mental")
).sort("date")

# get sum of all values by date
A = A.group_by("date").agg(pl.sum("valor"))

A = A[["date", "valor"]]

# linear regression
linear_regression = LinearRegression().fit(
    A["date"].cast(pl.Int64).to_numpy().reshape(-1, 1), A["valor"]
)

# poly regression
poly_features = PolynomialFeatures(degree=2).fit_transform(
    A["date"].cast(pl.Int64).to_numpy().reshape(-1, 1)
)
poly_regression = LinearRegression()
poly_regression.fit(poly_features, A["valor"])

# graph
plt.title("Ingresos Programa de Salud Mental")
plt.scatter(A["date"], A["valor"])
plt.plot(A["date"], linear_regression.predict(A["date"].cast(pl.Int64).to_numpy().reshape(-1, 1)), color="red")
plt.plot(A["date"], poly_regression.predict(poly_features), color="purple")
plt.show()
```

```python
# make graph for COSAM NUNOA
A = data_month.filter(
    pl.col("establecimiento_nombre") == "COSAM NUNOA",
    pl.col("reporte_descripcion").str.contains("Ingresos Programa de Salud Mental")
).sort("date")
A = A[["date", "valor"]]

dates = A["date"].to_numpy().reshape(-1, 1)

model = LinearRegression()
model.fit(A["date"].to_numpy().reshape(-1, 1), A["valor"])

predictions = model.predict(A["date"].cast(pl.Int64).to_numpy().reshape(-1, 1))

plt.title("Ingresos al programa de salud mental en COSAM NUNOA")
plt.scatter(A["date"], A["valor"])
plt.plot(A["date"], predictions, color="red")
plt.show()
```

```python
# look at "ingresos al programa" per commune, get linear regression for all communes, then sort by coefficient
communes = data_month.select("comuna_nombre").unique().to_series()

print(communes)
coefficients = []

for commune in communes:
    A = data_month.filter(
        pl.col("comuna_nombre") == commune,
        pl.col("reporte_descripcion").str.contains("Ingresos Programa de Salud Mental")
    ).sort("date")[["date", "valor"]]

    if(len(A) == 0):
        coefficients.append(None)
        continue

    model = LinearRegression()
    model.fit(A["date"].to_numpy().reshape(-1, 1), A["valor"])

    coefficients.append(model.coef_[0])

print(coefficients)
```

```python
# add coefficients to dataframe
coefficients_df = pl.DataFrame({
    "comuna_nombre": communes,
    "coef": coefficients
}).sort("coef", descending=True).filter(pl.col("coef").is_not_null())

coefficients_df
```

```python
# look at "ingresos al programa" per establishment, and make table of predictions sorted by most growth
establishments = data_month.select("establecimiento_nombre").unique().to_series()

coefficients = []

for establishment in establishments:
    A = data_month.filter(
        pl.col("establecimiento_nombre") == establishment,
        pl.col("reporte_descripcion").str.contains("Ingresos Programa de Salud Mental")
    ).sort("date")[["date", "valor"]]

    if(len(A) == 0):
        coefficients.append(None)
        continue

    # normalize
    A.with_columns(
        valor = MinMaxScaler().fit_transform(A["valor"].to_numpy().reshape(-1, 1))
    )

    model = LinearRegression()
    model.fit(A["date"].to_numpy().reshape(-1, 1), A["valor"])

    coefficients.append(model.coef_[0])

print(coefficients)
```

```python
# add coefficients to dataframe
coefficients_df = pl.DataFrame({
    "establishment": establishments,
    "coef": coefficients
}).sort("coef", descending=True).filter(pl.col("coef").is_not_null())

with pl.Config(fmt_str_lengths=1000, tbl_width_chars=1000):
    print(coefficients_df)
```
